Actually generating the bullets and writing them to files

```python
# Data Analysis
dataset_paths = 'bbc-news-dataset'
summaries_path = 'summaries'
bullets_path = 'bullets'


folders = os.listdir(dataset_paths + '/' + summaries_path)

for folder in folders:
    files = os.listdir(dataset_paths + '/' + summaries_path + '/' + folder)
    for file in files:
        summary_file_path = os.path.join(dataset_paths, summaries_path, folder, file)
        bullets_file_path = os.path.join(dataset_paths, bullets_path, folder, file)

        try:
            with open (summary_file_path,'r') as f:
                summary = f.read()
            response = None
            while response is None:
                try:
                    response = model.generate_content(summary)
                except Exception as e:
                    print(f"Quota likely exceeded. Trying again in 10 seconds. Error `{e}`")
                    response = None
                    sleep(10)
            with open (bullets_file_path,'w') as f:
                for bullet in json.loads(response.text)['bullets']:
                    f.write(bullet + '\n')
            print(f"Wrote bullet summary to {folder}/{file}", end="\r")
        except Exception as e:
            print(f"Error reading/writing file {folder}/{file} with error `{e}`")
```
